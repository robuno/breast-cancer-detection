```python
img_root_path = "./data/view_dset_01_splitted/cc_splitted/train/1/"
fixed_img_path = "./data/view_dset_01_splitted/cc_splitted/laterality_fixed/"
for img_path in os.listdir(img_root_path):
    img = Image.open(img_root_path + img_path)
    img = img.convert('L')
    img_array = np.asarray(img)
    img_array_T = np.transpose(img_array)
    if np.sum(img_array_T[0]) < np.sum(img_array_T[-1]):
        img = img.transpose(Image.FLIP_LEFT_RIGHT)
    img.save(fixed_img_path + img_path)
```
